```python
# o	Create a DataFrame with three columns: 'A', 'B', 'C'. 
# 'A' should contain 10 random integers, 'B' should contain 10 random floats, 
# and 'C' should contain 10 random strings.
A = np.random.randint(1, 100, size= 10)
B = np.random.rand(10)
C = [''.join(random.choices(string.ascii_uppercase, k=5)) for _ in range(10)]

d_frame = pd.DataFrame({'A': A, 'B': B, 'C': C})
```

```python
d_frame
```

```python
# o	Select the first 5 rows of the DataFrame.
d_frame.head()
```

```python
# o	Select the 'B' column from the DataFrame.
d_frame.B
```

```python
# o	Filter the DataFrame to show only rows where column 'A' is greater than 5.
d_frame[d_frame['A'] > 5]
```

```python
# o	Add a new column 'D' to the DataFrame, which is the product of 'A' and 'B'.
d_frame['D'] = d_frame['A'] * d_frame['B']
```

```python
#o	Rename the 'C' column to 'Strings'.
d_frame.rename(columns={'C':'Strings'}, inplace= True)
```
